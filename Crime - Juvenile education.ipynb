```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Dataset Description
```

The dataset contains education level of juveniles from 35 States/UT in India from 2000 to 2010

```python
df = pd.read_csv("18_01_Juveniles_arrested_Education.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df["Area_Name"].value_counts()
```

```python
df["Year"].value_counts()
```

```python
df["Year"] = df["Year"].astype(str)
df.dtypes
```

```python
df["Sub_Group_Name"].value_counts()
```

Clearly this column doesn't contain anything so let's drop it

```python
df.drop("Sub_Group_Name",axis=1,inplace=True)
df.head()
```

```python
df.shape
```

```python
state = df.groupby(['Area_Name'],sort = True).sum()
state
```

```python
t_state = state.transpose()
t_state
```

```python
plt.figure(figsize= (5,5))
plt.ylabel('Education level')
t_state["Bihar"].plot.bar()
```

```python
t_state.reset_index(level=0, inplace=True)
t_state
```

```python
t_state.rename(columns={"index": "Education Level"},inplace=True)
t_state
```

```python
t_state.sort_values('Bihar',inplace=True)
```

```python
t_state
```

```python
plt.figure(figsize= (10,8))
sns.barplot(x = t_state['Bihar'], y = t_state['Education Level'],order=t_state['Education Level'])
plt.xticks(rotation = 90) # slope of the words in the x axis 
plt.xlabel('Number of Juveniles in Bihar')
plt.ylabel('Education Level')
plt.title('Education Level vs Number of Juveniles in Bihar')
plt.show()
```
