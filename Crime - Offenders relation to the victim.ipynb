# Offender's relation to the victim

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("21_Offenders_known_to_the_victim.csv")
df
```

```python
df.shape
```

```python
df.info()
```

```python
Year = df.groupby(['Year'],sort = True).sum()
Year
```

```python
Year.plot(y=["known_persons", "Neighbours", "Other_Known_persons", "Parentsclose_family_members", "Relatives"], kind="bar",figsize= (22,10))
```

```python
state1 = df.groupby(['Area_Name'],sort = True).sum()
state1
```

```python
state1.plot(y=["known_persons", "Neighbours", "Other_Known_persons", "Parentsclose_family_members", "Relatives"], kind="bar",figsize= (25,25))
```
