Clearly the Sub_Group_Name feature has the same entries as Group_Name so let's drop the former

```python
df.drop("Sub_Group_Name",axis=1,inplace=True)
df.head()
```

Let's group the data as per states

```python
df.shape
```

```python
state = df.groupby(['Area_Name'],sort = True).sum()
state
```

Let's compare total number of cases in different states

```python
plt.figure(figsize= (22,10))
plt.ylabel('Total Number of Cases Reported')
state["K_A_Grand_Total"].plot.bar()
```

From the above plot it is clear that most number of kidnapping cases are reported from Uttar Pradesh

```python
group = df.groupby(['Group_Name'],sort = True).sum()
group
```

```python
plt.figure(figsize= (22,10))
plt.ylabel('Total Number of Cases Reported')
group["K_A_Grand_Total"].plot.bar()
```

From the above plot it is clear that the most common reason for kidnapping is for Marriage

```python
df.head()
```

```python
year = df.groupby(['Year'],sort = True).sum()
year
```

```python
plt.figure(figsize= (22,10))
plt.ylabel('Total Number of Cases Reported')
year["K_A_Cases_Reported"].plot.bar()
```

The number of kidnapping cases are increasing over the years
